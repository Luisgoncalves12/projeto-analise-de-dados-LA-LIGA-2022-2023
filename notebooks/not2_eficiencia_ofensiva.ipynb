```python
import pandas as pd

df = pd.read_csv("base.csv")
display(df)
```

```python
df_semgk = df[df["Position"].str.strip() != "GK"].copy()

df_semgk["GP"] = (df_semgk["Goals"] / df_semgk["MinP"]) * 90
df_semgk["AP"] = (df_semgk["Assist"] / df_semgk["MinP"]) * 90

df_900 = df_semgk[df_semgk["MinP"] > 900].copy()
```

```python
melhores_gols = df_900[["Player Name", "Team-name", "Goals", "MinP", "GP"]].sort_values("GP", ascending=False)
display(melhores_gols.head(10))
```

```python
melhores_assist = df_900[["Player Name", "Team-name", "Assist", "MinP", "AP"]].sort_values("AP", ascending=False)
display(melhores_assist.head(10))
```

```python
piores = df_900[["Player Name", "Team-name", "Goals", "Assist", "MinP", "GP", "AP"]].sort_values(["GP", "AP"], ascending=True)
display(piores.head(10))
```

```python
df_900[["Player Name", "Team-name", "Goals", "Assist", "MinP", "GP", "AP"]].to_excel("eficiencia_jogadores.xlsx", index=False)
```
