```python
import pandas as pd

df = pd.read_csv("base.csv")
display(df)
```

```python
idade_media = df.groupby('Team-name').agg(
    idade_media=('Age', 'mean'),
    total_gols=('Goals', 'sum'),
    total_assists=('Assist', 'sum')
)

# Salva o DataFrame ordenado de volta na variável
idade_media = idade_media.sort_values('idade_media', ascending=True)

display(idade_media)
```

```python
# 1. Cria o resumo e ordena
idade_media = df.groupby('Team-name').agg(
    idade_media=('Age', 'mean'),
    total_gols=('Goals', 'sum'),
    total_assists=('Assist', 'sum')
).sort_values('idade_media', ascending=False)

# 2. Seleciona apenas as colunas geradas (o Team-name já vai junto como índice)
# Mudamos para index=True para o nome do time aparecer no Excel
idade_media[["idade_media", "total_gols", "total_assists"]].to_excel("Idade_elenco.xlsx", index=True)

print("Arquivo salvo com sucesso!")
```
