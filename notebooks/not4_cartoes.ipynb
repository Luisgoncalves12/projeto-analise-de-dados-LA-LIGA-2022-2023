```python
import pandas as pd

df = pd.read_csv("base.csv")
display(df)
```

```python
df900 = df[df['MinP'] > 900]
d = df900.groupby('Team-name',as_index=False).agg(
    Total_Amarelos=('YC', 'sum'),
    Total_Vermelhos=('RC', 'sum')
)
display(d)
```

```python
d["Total Car"] = d["Total_Amarelos"] + d["Total_Vermelhos"]
d = d.sort_values("Total Car", ascending=False)
display(d)
```

```python
d.to_excel("Cartao.xlsx", index=False)
```
